### Step 1. Import the necessary libraries

```python
import pandas as pd

# package to extract data from various Internet sources into a DataFrame
# see https://pandas-datareader.readthedocs.io/en/latest/remote_data.html

import pandas_datareader.data as web

# package for dates
import datetime as dt
```

### Step 2. Create your time range (start and end variables). The start date should be 01/01/2015 and the end should today (whatever your today is)

```python
startdate = dt.datetime(2015, 1, 1)

enddate = dt.datetime.today()
```

### Step 3. Select the Apple, Tesla, Twitter, IBM, LinkedIn stocks symbols and assign them to a variable called stocks

```python
# replaced linkedin with microsoft as data for LNKD was not available
stocks = ['AAPL', 'TSLA', 'TWTR', 'IBM', 'MSFT']
```

### Step 4. Read the data from google, assign to df and print it

```python
# google not working so trying yahoo instead
df = web.DataReader(stocks, 'yahoo', start=startdate, end=enddate)
df
```

### Step 8. Aggregate the data of Volume to weekly
#### Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
# see https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects
# for time offset codes
vol.resample('W').sum()
```
