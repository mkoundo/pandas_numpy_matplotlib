### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
chipo
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
unit_price = chipo['item_price'] / chipo['quantity']
unit_price[unit_price > 10.0].count()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
item_name_price = chipo[['item_name', 'item_price']]
item_name_price
```

### Step 6. Sort by the name of the item

```python
item_name_price.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# combine series unit_price and column quantity into a dataframe
unit_price_quantity = pd.concat([unit_price, chipo['quantity']], axis=1)
unit_price_quantity.rename(columns = {0: 'unit_price'}, inplace=True)
unit_price_quantity
```

```python
# display quantity values for rows where unit_price = max unit price
filt = unit_price_quantity['unit_price'] == unit_price.max()
unit_price_quantity.loc[filt, 'quantity']
```

```python
# sum of quantity values for rows where unit_price = max unit price 
unit_price_quantity.loc[filt, 'quantity'].sum()
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
filt = chipo['item_name'] == 'Veggie Salad Bowl'
chipo.loc[filt, 'quantity'].sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
filt = (chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)
chipo.loc[filt]
```

```python
chipo.loc[filt]['quantity'].count()
```
